## MOS Capacitor

```python
mdl = BeckersQVpy()
mdl2 = DiracQVcpp()
mdl2.temp = mdl.temp = 30
```

```python
y, psi = mdl.y_psi(mdl.v_th)
y2, psi2 = mdl2.y_psi(mdl2.v_th)
mdl.v_th, mdl2.v_th
```

```python
plt.figure()
plt.plot(y*1e6, mdl.psi_v - psi, 'b')
plt.plot(y*1e6, mdl.psi_c - psi, 'b')
plt.plot(y2*1e6, mdl2.psi_v - psi2, 'g')
plt.plot(y2*1e6, mdl2.psi_c - psi2, 'g')
plt.axhline(0, c='r', ls='--')
plt.xlim(0, 0.2)
plt.xlabel('$y$ ($\mu$m)')
plt.ylabel('$\phi$ (V)')
plt.show()
```
