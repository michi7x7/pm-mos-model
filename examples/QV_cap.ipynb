## MOS Capacitor

Below are several plots for an NMOS-transistor in the threshold region ($n_{surface} = p_{bulk}$) for several temperatures. The oxide-bandgap is not to scale.

The default parameters are:
- $N_A$ = 1e23 / m$^3$
- $C_{ox}$ = 0.005755 F/m$^2$

```python
mdl = BeckersQVpy(temp = 300)
#mdl2 = DiracQVcpp()
```

```python
from CryMOS.plots import plot_moscap

for temp in [300, 200, 100, 50, 30]:
    mdl.temp = temp
    fig = plot_moscap(mdl, Eg_siox=2)
    fig.suptitle(f"NMOS in threshold condition ($V_{{GB}} = {mdl.v_th:.2f} V$) T={temp}K")
```
